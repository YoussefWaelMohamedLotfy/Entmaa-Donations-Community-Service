```python
import altair as alt
import numpy as np
import pandas as pd
import mpld3
from vega_datasets import data
from matplotlib import pyplot as plt
```

```python
data = pd.DataFrame({'a': np.random.rand(30),
                     'b': range(30)})
```

```python
chart = alt.Chart(data)
```

```python
chart.mark_area( line={'color':'darkgreen'},
    color=alt.Gradient(
        gradient='linear',
        stops=[alt.GradientStop(color='white', offset=0),
               alt.GradientStop(color='darkgreen', offset=1)],
        x1=1,
        x2=1,
        y1=1,
        y2=0
    )
    ).encode(y='a', x='b')
```
